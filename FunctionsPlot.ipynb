```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


# Spherical Function
def spherical(x, y):
    return x**2 + y**2

# Quadric Function
def quadric(x, y):
    return x**2 - y**2

# Ackley Function
def ackley(x, y):
    n = 2
    return -20 * np.exp(-0.2 * np.sqrt((1/n) * (x**2 + y**2))) - \
           np.exp((1/n) * (np.cos(2 * np.pi * x) + np.cos(2 * np.pi * y))) + np.exp(1) + 20

# Bohachevsky 1 Function
def bohachevsky1(x, y):
    return x**2 + 2 * y**2 - 0.3 * np.cos(3 * np.pi * x) * np.cos(4 * np.pi * y) + 0.3

# Colville Function
def colville(x, y):
    return 100 * (x**2 - y)**2 + (1 - x)**2 + 90 * (y - 1)**2 + 10.1 * (np.sin(3 * np.pi * x))**2 + 10.1 * (np.sin(3 * np.pi * y))**2 + 19.8 * np.sin(2 * np.pi * y)

# Easom Function
def easom(x, y):
    return -np.cos(x) * np.cos(y) * np.exp(-((x - np.pi)**2 + (y - np.pi)**2))

# Griewank Function
def griewank(x, y):
    return (x**2 + y**2) / 4000 - np.cos(x) * np.cos(y / np.sqrt(2)) + 1

# Hyperellipsoid Function
def hyperellipsoid(x, y):
    return x**2 + 2 * y**2

# Rastrigin Function
def rastrigin(x, y):
    return 20 + x**2 - 10 * np.cos(2 * np.pi * x) + y**2 - 10 * np.cos(2 * np.pi * y)

# Rosenbrock Function
def rosenbrock(x, y):
    return (1 - x)**2 + 100 * (y - x**2)**2

# Schwefel Function
def schwefel(x, y):
    return -x * np.sin(np.sqrt(np.abs(x))) - y * np.sin(np.sqrt(np.abs(y)))

# Create a grid of values for x and y
x100 = np.linspace(-100, 100, 100)
y100 = np.linspace(-100, 100, 100)
X100, Y100 = np.meshgrid(x100, y100)

x30 = np.linspace(-30, 30, 100)
y30 = np.linspace(-30, 30, 100)
X30, Y30 = np.meshgrid(x30, y30)

x50 = np.linspace(-50, 50, 100)
y50 = np.linspace(-50, 50, 100)
X50, Y50 = np.meshgrid(x50, y50)

x10 = np.linspace(-10, 10, 100)
y10 = np.linspace(-10, 10, 100)
X10, Y10 = np.meshgrid(x10, y10)

x600 = np.linspace(-600, 600, 100)
y600 = np.linspace(-600, 600, 100)
X600, Y600 = np.meshgrid(x600, y600)

x1 = np.linspace(-1, 1, 100)
y1 = np.linspace(-1, 1, 100)
X1, Y1 = np.meshgrid(x1, y1)

x512 = np.linspace(-5.12, 5.12, 100)
y512 = np.linspace(-5.12, 5.12, 100)
X512, Y512 = np.meshgrid(x512, y512)

x2048 = np.linspace(-2.048, 2.048, 100)
y2048 = np.linspace(-2.048, 2.048, 100)
X2048, Y2048 = np.meshgrid(x2048, y2048)

x500 = np.linspace(-500, 500, 100)
y500 = np.linspace(-500, 500, 100)
X500, Y500 = np.meshgrid(x500, y500)

# Evaluate the functions on the grid
Z_spherical = spherical(X100, Y100)
Z_quadric = quadric(X100, Y100)
Z_ackley = ackley(X30, Y30)
Z_bohachevsky1 = bohachevsky1(X50, Y50)
Z_colville = colville(X10, Y10)
Z_easom = easom(X100, Y100)
Z_griewank = griewank(X600, Y600)
Z_hyperellipsoid = hyperellipsoid(X1, Y1)
Z_rastrigin = rastrigin(X512, Y512)
Z_rosenbrock = rosenbrock(X2048, Y2048)
Z_schwefel = schwefel(X500, Y500)

# Create 3D plots for each function
functions = [
    (Z_spherical, 'Spherical'),
    (Z_quadric, 'Quadric'),
    (Z_ackley, 'Ackley'),
    (Z_bohachevsky1, 'Bohachevsky 1'),
    (Z_colville, 'Colville'),
    (Z_easom, 'Easom'),
    (Z_griewank, 'Griewank'),
    (Z_hyperellipsoid, 'Hyperellipsoid'),
    (Z_rastrigin, 'Rastrigin'),
    (Z_rosenbrock, 'Rosenbrock'),
    (Z_schwefel, 'Schwefel')
]

for Z, name in functions:
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(X1, Y1, Z, cmap='viridis')
    ax.set_title(name)

plt.show()
```
